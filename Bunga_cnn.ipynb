```python


import os
import random
import shutil
import json
import pickle
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import kagglehub
import splitfolders
import seaborn as sns

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from tensorflow.keras import layers, models, callbacks, regularizers
from tensorflow.keras.preprocessing import image_dataset_from_directory
from tensorflow.keras.utils import load_img, img_to_array, save_img
from google.colab import files

print("TensorFlow version:", tf.__version__)

SEED = 42
tf.random.set_seed(SEED)
np.random.seed(SEED)
random.seed(SEED)

# =========================
# 1. DOWNLOAD DATASET CIFAR-100
# =========================
path = kagglehub.dataset_download("fedesoriano/cifar100")
print("Path dataset:", path)
print("Isi root:", os.listdir(path))

# Struktur dataset langsung:
# file.txt, meta, train, test
CIFAR_DIR = path

meta_path = os.path.join(CIFAR_DIR, "meta")
train_path = os.path.join(CIFAR_DIR, "train")
test_path = os.path.join(CIFAR_DIR, "test")

required_files = [meta_path, train_path, test_path]
for file_path in required_files:
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"File tidak ditemukan: {file_path}")

print("Menggunakan folder CIFAR-100:", CIFAR_DIR)
print("File meta :", meta_path)
print("File train:", train_path)
print("File test :", test_path)

# =========================
# 2. FUNGSI LOAD PICKLE
# =========================
def unpickle(file):
    with open(file, "rb") as fo:
        data = pickle.load(fo, encoding="bytes")
    return data

meta = unpickle(meta_path)
train_dict = unpickle(train_path)
test_dict = unpickle(test_path)

fine_label_names = [x.decode("utf-8") for x in meta[b"fine_label_names"]]
coarse_label_names = [x.decode("utf-8") for x in meta[b"coarse_label_names"]]

print("Jumlah fine labels:", len(fine_label_names))
print("Jumlah coarse labels:", len(coarse_label_names))
print("Contoh fine labels bunga yang benar:",
      [lbl for lbl in fine_label_names if lbl in ["orchid", "poppy", "rose", "sunflower", "tulip"]])

# =========================
# 3. PILIH KELAS BUNGA YANG BENAR
# =========================
flower_classes = ["orchid", "poppy", "rose", "sunflower", "tulip"]

for cls in flower_classes:
    if cls not in fine_label_names:
        raise ValueError(f"Kelas {cls} tidak ditemukan di CIFAR-100")

flower_indices = {cls: fine_label_names.index(cls) for cls in flower_classes}
print("Index kelas bunga:", flower_indices)

# =========================
# 4. HITUNG JUMLAH DATA PER KELAS
# =========================
train_fine_labels = np.array(train_dict[b"fine_labels"])
test_fine_labels = np.array(test_dict[b"fine_labels"])

class_counts = {}
for cls, idx in flower_indices.items():
    count_train = np.sum(train_fine_labels == idx)
    count_test = np.sum(test_fine_labels == idx)
    class_counts[cls] = {
        "train": int(count_train),
        "test": int(count_test),
        "total": int(count_train + count_test)
    }

print("Jumlah data tiap kelas bunga:")
for cls, cnt in class_counts.items():
    print(f"{cls}: train={cnt['train']}, test={cnt['test']}, total={cnt['total']}")

# Ambil 4 kelas
selected_classes = sorted(class_counts.keys(), key=lambda x: (-class_counts[x]["total"], x))[:4]
selected_indices = [flower_indices[cls] for cls in selected_classes]

print("4 kelas terpilih:", selected_classes)

# =========================
# 5. SIAPKAN FOLDER OUTPUT
# =========================
RAW_DIR = "cifar100_flowers_raw"
TEMP_SPLIT_DIR = "temp_train_split"
SUBMISSION_DIR = "submission"

for folder in [RAW_DIR, TEMP_SPLIT_DIR, SUBMISSION_DIR]:
    if os.path.exists(folder):
        shutil.rmtree(folder)

os.makedirs(RAW_DIR, exist_ok=True)

RAW_TRAIN_DIR = os.path.join(RAW_DIR, "train")
RAW_TEST_DIR = os.path.join(RAW_DIR, "test")

for cls in selected_classes:
    os.makedirs(os.path.join(RAW_TRAIN_DIR, cls), exist_ok=True)
    os.makedirs(os.path.join(RAW_TEST_DIR, cls), exist_ok=True)

# =========================
# 6. KONVERSI ARRAY CIFAR -> GAMBAR FILE
# =========================
def cifar_row_to_image(row):
    img = row.reshape(3, 32, 32).transpose(1, 2, 0)
    return img

train_data = train_dict[b"data"]
test_data = test_dict[b"data"]

train_counter = {cls: 0 for cls in selected_classes}
for i in range(len(train_data)):
    label_idx = train_fine_labels[i]
    if label_idx in selected_indices:
        class_name = fine_label_names[label_idx]
        img = cifar_row_to_image(train_data[i])
        fname = f"{class_name}_{train_counter[class_name]:04d}.png"
        save_path = os.path.join(RAW_TRAIN_DIR, class_name, fname)
        save_img(save_path, img)
        train_counter[class_name] += 1

test_counter = {cls: 0 for cls in selected_classes}
for i in range(len(test_data)):
    label_idx = test_fine_labels[i]
    if label_idx in selected_indices:
        class_name = fine_label_names[label_idx]
        img = cifar_row_to_image(test_data[i])
        fname = f"{class_name}_{test_counter[class_name]:04d}.png"
        save_path = os.path.join(RAW_TEST_DIR, class_name, fname)
        save_img(save_path, img)
        test_counter[class_name] += 1

print("Data train per kelas setelah ekstraksi:", train_counter)
print("Data test per kelas setelah ekstraksi:", test_counter)

# =========================
# 7. SPLIT TRAIN -> TRAIN/VAL
# TEST TETAP DARI TEST ASLI
# =========================
splitfolders.ratio(
    RAW_TRAIN_DIR,
    output=TEMP_SPLIT_DIR,
    seed=SEED,
    ratio=(0.8, 0.2)
)

TRAIN_DIR = os.path.join(TEMP_SPLIT_DIR, "train")
VAL_DIR = os.path.join(TEMP_SPLIT_DIR, "val")
TEST_DIR = RAW_TEST_DIR

print("Train directory:", TRAIN_DIR)
print("Val directory:", VAL_DIR)
print("Test directory:", TEST_DIR)

# =========================
# 8. LOAD DATASET TF
# =========================
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE

train_ds_raw = image_dataset_from_directory(
    TRAIN_DIR,
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_ds_raw = image_dataset_from_directory(
    VAL_DIR,
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

test_ds_raw = image_dataset_from_directory(
    TEST_DIR,
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    shuffle=False
)

class_names = train_ds_raw.class_names
num_classes = len(class_names)

print("Class names:", class_names)
print("Number of classes:", num_classes)

def count_images(folder):
    total = 0
    for root, dirs, files in os.walk(folder):
        total += len([f for f in files if f.lower().endswith((".jpg", ".jpeg", ".png"))])
    return total

train_count = count_images(TRAIN_DIR)
val_count = count_images(VAL_DIR)
test_count = count_images(TEST_DIR)

print("Train images:", train_count)
print("Val images:", val_count)
print("Test images:", test_count)
print("Total images:", train_count + val_count + test_count)

# =========================
# 9. VISUALISASI SAMPLE
# =========================
plt.figure(figsize=(12, 8))
for images, labels in train_ds_raw.take(1):
    for i in range(min(9, len(images))):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
plt.tight_layout()
plt.show()

# =========================
# 10. AUGMENTASI & PREPROCESS
# =========================
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.2),
    layers.RandomZoom(0.2),
    layers.RandomContrast(0.2),
    layers.RandomTranslation(0.1, 0.1)
])

preprocess_input = tf.keras.applications.mobilenet_v2.preprocess_input

def preprocess_train(x, y):
    x = data_augmentation(x, training=True)
    x = preprocess_input(x)
    return x, y

def preprocess_eval(x, y):
    x = preprocess_input(x)
    return x, y

train_ds = train_ds_raw.map(preprocess_train, num_parallel_calls=AUTOTUNE).cache().shuffle(1000).prefetch(AUTOTUNE)
val_ds = val_ds_raw.map(preprocess_eval, num_parallel_calls=AUTOTUNE).cache().prefetch(AUTOTUNE)
test_ds = test_ds_raw.map(preprocess_eval, num_parallel_calls=AUTOTUNE).cache().prefetch(AUTOTUNE)

## =========================
# 11. MODEL
# =========================
base_model = tf.keras.applications.MobileNetV2(
    input_shape=(224, 224, 3),
    include_top=False,
    weights="imagenet",
    pooling="avg"
)

base_model.trainable = False

model = models.Sequential([
    layers.Input(shape=(224, 224, 3)),
    base_model,
    layers.BatchNormalization(),
    layers.Dense(256, activation="relu", kernel_regularizer=regularizers.l2(0.001)),
    layers.Dropout(0.5),
    layers.Dense(num_classes, activation="softmax")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()

# =========================
# 12. CALLBACKS
# =========================
early_stop = callbacks.EarlyStopping(
    monitor="val_accuracy",
    patience=6,
    restore_best_weights=True
)

reduce_lr = callbacks.ReduceLROnPlateau(
    monitor="val_loss",
    factor=0.5,
    patience=2,
    verbose=1
)

checkpoint = callbacks.ModelCheckpoint(
    "best_model_stage1.keras",
    monitor="val_accuracy",
    save_best_only=True,
    verbose=1
)

class StopAtAccuracy(callbacks.Callback):
    def on_epoch_end(self, epoch, logs=None):
        logs = logs or {}
        train_acc = logs.get("accuracy", 0)
        val_acc = logs.get("val_accuracy", 0)
        if train_acc >= 0.85 and val_acc >= 0.85:
            print("\nTarget accuracy reached. Stopping training.")
            self.model.stop_training = True

stop_at_acc = StopAtAccuracy()

# =========================
# 13. TRAINING STAGE 1
# =========================
history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=15,
    callbacks=[early_stop, reduce_lr, checkpoint, stop_at_acc]
)

# =========================
# 14. FINE TUNING
# =========================
base_model.trainable = True

fine_tune_at = 100
for layer in base_model.layers[:fine_tune_at]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

checkpoint_ft = callbacks.ModelCheckpoint(
    "best_model_finetune.keras",
    monitor="val_accuracy",
    save_best_only=True,
    verbose=1
)

history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=25,
    initial_epoch=len(history1.history["accuracy"]),
    callbacks=[early_stop, reduce_lr, checkpoint_ft]
)
# =========================
# 15. EVALUASI
# =========================
train_loss, train_acc = model.evaluate(train_ds, verbose=1)
val_loss, val_acc = model.evaluate(val_ds, verbose=1)
test_loss, test_acc = model.evaluate(test_ds, verbose=1)

print(f"Train Accuracy : {train_acc*100:.2f}%")
print(f"Val Accuracy   : {val_acc*100:.2f}%")
print(f"Test Accuracy  : {test_acc*100:.2f}%")

acc = history1.history["accuracy"] + history2.history["accuracy"]
val_acc_hist = history1.history["val_accuracy"] + history2.history["val_accuracy"]
loss = history1.history["loss"] + history2.history["loss"]
val_loss_hist = history1.history["val_loss"] + history2.history["val_loss"]

epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(10, 5))
plt.plot(epochs_range, acc, label="Training Accuracy")
plt.plot(epochs_range, val_acc_hist, label="Validation Accuracy")
plt.title("Training and Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.plot(epochs_range, loss, label="Training Loss")
plt.plot(epochs_range, val_loss_hist, label="Validation Loss")
plt.title("Training and Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()

# =========================
# 16. EXPORT SAVEDMODEL + TFLITE
# =========================
saved_model_path = "saved_model"
if os.path.exists(saved_model_path):
    shutil.rmtree(saved_model_path)

model.export(saved_model_path)
print(f"SavedModel saved to: {saved_model_path}")

converter = tf.lite.TFLiteConverter.from_saved_model(saved_model_path)
tflite_model = converter.convert()

with open("model.tflite", "wb") as f:
    f.write(tflite_model)

print("TFLite model saved to: model.tflite")

# =========================
# 17. CONTOH PREDIKSI 1 GAMBAR DARI TEST
# =========================
for test_images, test_labels in test_ds_raw.take(1):
    sample_img = test_images[0]
    sample_label = test_labels[0].numpy()
    break

img_array = tf.expand_dims(sample_img, axis=0)
img_input = preprocess_input(tf.cast(img_array, tf.float32))

pred = model.predict(img_input, verbose=0)
pred_index = np.argmax(pred[0])

print("Ground Truth :", class_names[sample_label])
print("Prediction   :", class_names[pred_index])
print("Confidence   :", float(np.max(pred[0])))

plt.figure(figsize=(4, 4))
plt.imshow(sample_img.numpy().astype("uint8"))
plt.title(f"True: {class_names[sample_label]}\nPred: {class_names[pred_index]}")
plt.axis("off")
plt.show()

# =========================
# 18. UJI MODEL TFLITE
# =========================
interpreter = tf.lite.Interpreter(model_path="model.tflite")
interpreter.allocate_tensors()

input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

input_data = tf.cast(img_input, dtype=input_details[0]["dtype"]).numpy()
interpreter.set_tensor(input_details[0]["index"], input_data)
interpreter.invoke()

tflite_pred = interpreter.get_tensor(output_details[0]["index"])
tflite_pred_index = np.argmax(tflite_pred[0])

print("TFLite Prediction :", class_names[tflite_pred_index])
print("TFLite Confidence :", float(np.max(tflite_pred[0])))

# =========================
# 19. CONFUSION MATRIX + REPORT
# =========================
y_true = np.concatenate([y.numpy() for x, y in test_ds], axis=0)
y_pred_probs = model.predict(test_ds, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)

test_acc_manual = accuracy_score(y_true, y_pred)
print(f"Akurasi test (manual) : {test_acc_manual*100:.2f}%")
print(classification_report(y_true, y_pred, target_names=class_names))

cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.title("Confusion Matrix")
plt.show()

# =========================
# 20. HITUNG PREDIKSI BENAR / SALAH
# =========================
all_images = []
all_labels = []

for images, labels in test_ds_raw:
    all_images.append(images.numpy())
    all_labels.append(labels.numpy())

all_images = np.concatenate(all_images, axis=0)
all_labels = np.concatenate(all_labels, axis=0)

correct_idx = np.where(y_pred == all_labels)[0]
wrong_idx = np.where(y_pred != all_labels)[0]

print(f"Jumlah prediksi benar : {len(correct_idx)}")
print(f"Jumlah prediksi salah : {len(wrong_idx)}")

# =========================
# 21. PREDIKSI GAMBAR UPLOAD
# =========================
uploaded = files.upload()
img_path = next(iter(uploaded))

img = load_img(img_path, target_size=(224, 224))
img_array = img_to_array(img)
img_input = preprocess_input(img_array.copy())
img_input = np.expand_dims(img_input, axis=0)

pred = model.predict(img_input, verbose=0)
pred_idx = np.argmax(pred[0])
confidence = float(np.max(pred[0]))

plt.figure(figsize=(5, 5))
plt.imshow(img)
plt.title(f"Pred: {class_names[pred_idx]}\nConfidence: {confidence:.4f}")
plt.axis("off")
plt.show()

print("Prediction :", class_names[pred_idx])
print("Confidence :", confidence)

# =========================
# 22. SIAPKAN FOLDER SUBMISSION
# =========================
TFJS_DIR = os.path.join(SUBMISSION_DIR, "tfjs_model")
TFLITE_DIR = os.path.join(SUBMISSION_DIR, "tflite")
SAVED_MODEL_DIR = os.path.join(SUBMISSION_DIR, "saved_model")

os.makedirs(TFJS_DIR, exist_ok=True)
os.makedirs(TFLITE_DIR, exist_ok=True)
os.makedirs(SAVED_MODEL_DIR, exist_ok=True)

model.save(os.path.join(SUBMISSION_DIR, "klasifikasi-bunga-cifar100.keras"))
model.export(SAVED_MODEL_DIR)

converter = tf.lite.TFLiteConverter.from_saved_model(SAVED_MODEL_DIR)
tflite_model = converter.convert()

with open(os.path.join(TFLITE_DIR, "model.tflite"), "wb") as f:
    f.write(tflite_model)

with open(os.path.join(TFLITE_DIR, "label.txt"), "w") as f:
    for label in class_names:
        f.write(label + "\n")

!tensorflowjs_converter --input_format=tf_saved_model "{SAVED_MODEL_DIR}" "{TFJS_DIR}"

readme_content = f"""# Image Classification Submission

## Nama Proyek
Klasifikasi 4 Kelas Bunga CIFAR-100

## Dataset
CIFAR-100 via KaggleHub

## Kelas
{", ".join(class_names)}

## Arsitektur Model
Sequential + Conv2D + MaxPooling2D + MobileNetV2

## Akurasi
- Train Accuracy  : {train_acc*100:.2f}%
- Val Accuracy    : {val_acc*100:.2f}%
- Test Accuracy   : {test_acc*100:.2f}%
"""

with open(os.path.join(SUBMISSION_DIR, "README.md"), "w", encoding="utf-8") as f:
    f.write(readme_content)

requirements_content = """tensorflow
kagglehub
split-folders
numpy
matplotlib
scikit-learn
seaborn
tensorflowjs
"""

with open(os.path.join(SUBMISSION_DIR, "requirements.txt"), "w", encoding="utf-8") as f:
    f.write(requirements_content)

placeholder_notebook = {
    "cells": [
        {
            "cell_type": "markdown",
            "metadata": {},
            "source": [
                "# Notebook Submission\n",
                "Ganti file ini dengan notebook asli .ipynb.\n"
            ]
        }
    ],
    "metadata": {},
    "nbformat": 4,
    "nbformat_minor": 5
}

with open(os.path.join(SUBMISSION_DIR, "notebook.ipynb"), "w", encoding="utf-8") as f:
    json.dump(placeholder_notebook, f, indent=2)

for root, dirs, files in os.walk(SUBMISSION_DIR):
    level = root.replace(SUBMISSION_DIR, "").count(os.sep)
    indent = " " * 4 * level
    print(f"{indent}{os.path.basename(root)}/")
    subindent = " " * 4 * (level + 1)
    for file in files:
        print(f"{subindent}{file}")
```
